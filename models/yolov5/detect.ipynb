# Litter Detection in Glasgow City using Yolov5

## Configuration

```python
model = "mosaic-noise-x3"
weights = f"yolov5/runs/train/{model}/weights/best.pt"

images = "../../data/images"
source = f"{images}/../images-expanded"
destination = "yolov5/runs/detect"

confidence_threshold = 0.25
line_thickness = 3
```

_Optional:_ Create the source directory that contains all of the images to run inference on.

```python
import glob
import shutil
import os
import filecmp

remove_dups = False

if not os.path.isdir(source):
    os.makedirs(source)
    files = glob.iglob(f"{images}/**/*.jpg", recursive=True)
    for f in files:
        dup = False
        for g in os.listdir(source):
            if remove_dups and filecmp.cmp(f, os.path.join(source, g), shallow=False):
                dup = True
                break
        if not dup:
            shutil.copy(f, source)
```

## Inference

```python
%%time
!python yolov5/detect.py --weights $weights --source $source --project $destination --name $model --conf $confidence_threshold --line-thickness $line_thickness --save-txt --save-conf --hide-labels --hide-conf
```
