# Litter Detection in Glasgow City using Yolov5

## Configuration

```python
model = "mosaic-noise-x3"
weights = f"yolov5/runs/train/{model}/weights/best.pt"

images = "../../data/images"
source = f"{images}/../images-expanded"
destination = "yolov5/runs/detect"

confidence_threshold = 0.5
line_thickness = 3
```

_Optional:_ Create the source directory that contains all of the images to run inference on.

```python
import glob
import shutil
import os

if not os.path.isdir(source):
    os.makedirs(source)
    for f in glob.iglob(f"{images}/**/*.jpg", recursive=True):
        shutil.copy(f, source)
```

## Inference

```python
%%time
!python yolov5/detect.py --weights $weights --source $destination/images --project $destination --name $model --conf $confidence_threshold --line-thickness $line_thickness --save-txt --save-conf --hide-labels --hide-conf
```
